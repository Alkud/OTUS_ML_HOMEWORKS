```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
data_file_name = 'wdbc.data'
```

```python
data_header = ['ID', 'Diagnosis']
header_base_3to32 = ['Radius', 'Texture', 'Perimeter', 'Area', 'Smoothness', 'Compactness', 'Concavity', 'Concave_points', 'Symmetry', 'Fractal_dimension']
subheadres_3to32 = ['mean', 'std', 'worst']
headers_3to32 = []
for base in header_base_3to32:
    for subheader in subheadres_3to32:
        headers_3to32.append('_'.join([base,subheader]))
data_header.extend(headers_3to32)
len(data_header)
```

```python
data_frame = pd.read_csv(data_file_name, names = data_header).fillna('')
```

```python
data_frame.info()
```

```python
data_frame.head()
```

```python
data_frame.describe()
```

```python
corr_matrix = data_frame.corr()
```

```python
most_correlated_variables = []
for label_1 in corr_matrix.index:
    for label_2 in corr_matrix.columns[np.abs((corr_matrix[label_1] > 0.95)) & np.abs((corr_matrix[label_1]) < 1.0)]:        
        if label_1 not in most_correlated_variables:
            most_correlated_variables.append(label_1)            
        if label_2 not in most_correlated_variables:
            most_correlated_variables.append(label_2)
            print (label_1, '--', label_2, ':', corr_matrix[label_1][label_2])
```

```python
most_correlated_variables.sort()
most_correlated_variables
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(corr_matrix.loc[most_correlated_variables, most_correlated_variables], annot=True, fmt='.2f', center = 0.5)
```

```python
M_group = data_frame.loc[data_frame['Diagnosis'] == 'M']
B_group = data_frame.loc[data_frame['Diagnosis'] == 'B']
```

```python
M_group.info()
```

```python
B_group.info()
```

```python
from scipy import stats
```

```python
#selected_features = ['Concave_points_std', 'Radius_mean', 'Texture_mean', 'Radius_std', 'Compactness_std', 'Area_mean']
selected_features = data_header.copy()
selected_features.remove('ID')
selected_features.remove('Diagnosis')

MB_p_values = []

for feature in selected_features:
    t, p_value = stats.ttest_ind(M_group[feature], B_group[feature])
    MB_p_values.append(p_value)
    print(f'Различие средних значений признака {feature} для диагнозов M и B')
    if p_value < 0.05:
        print(f'\tстатистически значимое, p_value={p_value}.')
    else:
        print(f'\t\tстатистически НЕзначимое, p_value={p_value}.')
```

```python
encoded_data = data_frame.drop('ID', axis=1)
encoded_data['Diagnosis'] = preprocessing.LabelEncoder().fit_transform(encoded_data['Diagnosis'])
encoded_data.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
y = encoded_data.Diagnosis
X = encoded_data.drop('Diagnosis', axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_train = min_max_scaler.fit_transform(X_train)
X_test = min_max_scaler.transform (X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf_regressor = RandomForestRegressor(max_depth=2, random_state=0, n_estimators=100)
rf_regressor.fit(X_train, y_train)
```

```python
importances = list(zip(rf_regressor.feature_importances_, MB_p_values, X.columns))
importances.sort(reverse=True)
```
