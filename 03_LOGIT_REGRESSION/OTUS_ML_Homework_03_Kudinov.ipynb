# Составим список наиболее коррелированных параметров и удалим их из набора   
___

```python
corr_matrix = data_frame.corr()
most_correlated_variables = []
for label_1 in corr_matrix.index:
    for label_2 in corr_matrix.columns[np.abs((corr_matrix[label_1] >= 0.85)) & np.abs((corr_matrix[label_1]) < 1.0)]:        
        if label_1 not in most_correlated_variables:
            most_correlated_variables.append(label_1)            
        if label_2 not in most_correlated_variables:
            most_correlated_variables.append(label_2)
            print (label_1, '--', label_2, ':', corr_matrix[label_1][label_2])
```

```python
most_correlated_variables.sort()
most_correlated_variables
```

```python
data_frame = data_frame.drop(most_correlated_variables, axis=1)
data_frame = data_frame.drop(['id'], axis=1)
```

___
# Для оставшихся параметров построим распредления значений, сопровождающих один из двух диагнозов
___

```python
corr_matrix = data_frame.corr()
fairly_correlated_variables = corr_matrix.mean()[corr_matrix.mean()<=0.45].index.tolist()
strongly_correlated_variables = corr_matrix.mean()[corr_matrix.mean()>0.45].index.tolist()
```

# Перекодируем значение целевой переменной:  
# M -  1 (злокачественная), B - 0 (доброкачественная)
___


```python
data_frame['diagnosis'] = data_frame['diagnosis'].apply(lambda x: 1 if x=='M' else 0)
```

___
# Разбиваем набор на тренировочные и тестовые данные.
# Масштабируем значения всех параметров так, чтобы их средния значения были равны 0, а дисперсия была равна 1:
___

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    data_frame.drop(['diagnosis'], axis=1), data_frame['diagnosis'],
    test_size=0.3, random_state=1812, stratify=data_frame['diagnosis']
)
```

```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
def quality_report(prediction, actual):
    print("Accuracy: {:.3f}\nPrecision: {:.3f}\nRecall: {:.3f}\nf1_score: {:.3f}".format(
        accuracy_score(prediction, actual),
        precision_score(prediction, actual),
        recall_score(prediction, actual),
        f1_score(prediction, actual)
    ))
```

```python
from sklearn.linear_model import LogisticRegression

simple_model = LogisticRegression()

simple_model.fit(X_train_scaled, y_train)
print("Train quality:")
quality_report(simple_model.predict(X_train_scaled), y_train)
print("\nTest quality:")
quality_report(simple_model.predict(X_test_scaled), y_test)
```

```python
from sklearn.metrics import roc_auc_score, roc_curve

def plot_roc_curve(prob_prediction, actual):
    fpr, tpr, thresholds = roc_curve(actual, prob_prediction)
    auc_score = roc_auc_score(actual, prob_prediction)
    
    plt.plot(fpr, tpr, label='ROC curve ')
    plt.plot([0, 1], [0, 1])
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC AUC: {:.3f}'.format(auc_score))
    plt.show()
```

```python
plot_roc_curve(simple_model.predict_proba(X_test_scaled)[:,1], y_test)
```

```python
featureImportance = pd.DataFrame({"feature": data_frame.drop('diagnosis',axis=1).columns, 
                                  "importance": simple_model.coef_[0]})

featureImportance.set_index('feature', inplace=True)
featureImportance.sort_values(['importance'], ascending=False, inplace=True)
plot_colors = ['b' if coeff<0 else 'r' for coeff in featureImportance['importance'].tolist()]
featureImportance['importance'].plot('bar', figsize=(10, 10), color=plot_colors);
```

```python
from sklearn.linear_model import LogisticRegressionCV
```

```python
cv_model = LogisticRegressionCV(Cs=100, cv=4)
cv_model.fit(X_train_scaled, y_train)
print("Train quality:")
quality_report(cv_model.predict(X_train_scaled), y_train)
print("\nTest quality:")
quality_report(cv_model.predict(X_test_scaled), y_test)
```

```python
plot_roc_curve(cv_model.predict_proba(X_test_scaled)[:,1], y_test)
```

```python
featureImportance = pd.DataFrame({"feature": data_frame.drop('diagnosis',axis=1).columns, 
                                  "importance": cv_model.coef_[0]})

featureImportance.set_index('feature', inplace=True)
featureImportance.sort_values(['importance'], ascending=False, inplace=True)
plot_colors = ['b' if coeff<0 else 'r' for coeff in featureImportance['importance'].tolist()]
featureImportance['importance'].plot('bar', figsize=(10, 10), color=plot_colors);
```

___
# Кросс-валидация подтвердила достоверность метрик качества простой модели. Но значения весов уточнились.  
# Проверим, не отбросили ли мы важные параметры. Из пар наиболее коррелированных переменных (выводился в ячейке #17) оставим по одной: radius_mean, compactness_mean, radius_se
___

```python
reject_list = most_correlated_variables
reject_list.append('id')
reject_list.remove('radius_mean')
reject_list.remove('compactness_mean')
reject_list.remove('radius_se')
```

```python
full_data_frame = data_frame = pd.read_csv(data_file_name).drop('Unnamed: 32', axis=1)
full_data_frame = full_data_frame.drop(reject_list, axis=1)

full_data_frame['diagnosis'] = full_data_frame['diagnosis'].apply(lambda x: 1 if x=='M' else 0)

X_train, X_test, y_train, y_test = train_test_split(
    full_data_frame.drop(['diagnosis'], axis=1), full_data_frame['diagnosis'],
    test_size=0.3, random_state=1812, stratify=full_data_frame['diagnosis']
)

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

simple_model.fit(X_train_scaled, y_train)
print("Train quality:")
quality_report(simple_model.predict(X_train_scaled), y_train)
print("\nTest quality:")
quality_report(simple_model.predict(X_test_scaled), y_test)
```

```python
cv_model.fit(X_train_scaled, y_train)
print("Train quality:")
quality_report(cv_model.predict(X_train_scaled), y_train)
print("\nTest quality:")
quality_report(cv_model.predict(X_test_scaled), y_test)
```

```python
featureImportance = pd.DataFrame({"feature": full_data_frame.drop('diagnosis',axis=1).columns, 
                                  "importance": cv_model.coef_[0]})

featureImportance.set_index('feature', inplace=True)
featureImportance.sort_values(['importance'], ascending=False, inplace=True)
plot_colors = ['b' if coeff<0 else 'r' for coeff in featureImportance['importance'].tolist()]
featureImportance['importance'].plot('bar', figsize=(10, 10), color=plot_colors);
```

___
# С возвратом ранее отброшенных параметров точность классификации значительно выросла. Более того, ранее отброшенные параметры входят в число наиболее важных.
___

# Сравним точность нашей модели с точностью модели KNN.
___

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

param_grid = {'n_neighbors': range(1, 50), 'p': range(1, 4)}

knn_model = KNeighborsClassifier()

grid_search = GridSearchCV(knn_model, param_grid, cv=5, scoring='f1')
grid_search.fit(X_train_scaled, y_train)

print("Best CV score: {:.3f}, best CV k: {}".format(
    grid_search.best_score_, grid_search.best_estimator_.n_neighbors)
)
```

```python
y_pred = grid_search.best_estimator_.predict(X_test_scaled)
quality_report(y_pred, y_test)
```
